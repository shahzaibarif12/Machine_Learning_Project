```python
from matplotlib import pyplot
from pandas import read_csv
import pandas as pd
filename = 'facial_keypoints.csv'
names = ['left_eye_center_x','left_eye_center_y','right_eye_center_x','right_eye_center_y','left_eye_inner_corner_x','left_eye_inner_corner_y','left_eye_outer_corner_x','left_eye_outer_corner_y','right_eye_inner_corner_x','right_eye_inner_corner_y','right_eye_outer_corner_x','right_eye_outer_corner_y','left_eyebrow_inner_end_x','left_eyebrow_inner_end_y','left_eyebrow_outer_end_x','left_eyebrow_outer_end_y','right_eyebrow_inner_end_x','right_eyebrow_inner_end_y','right_eyebrow_outer_end_x','right_eyebrow_outer_end_y','nose_tip_x','nose_tip_y','mouth_left_corner_x','mouth_left_corner_y','mouth_right_corner_x','mouth_right_corner_y','mouth_center_top_lip_x','mouth_center_top_lip_y','mouth_center_bottom_lip_x','mouth_center_bottom_lip_y']
data= read_csv(filename, names=names, delimiter=",", header=None)

data['left_eye_center_x']= pd.to_numeric(data['left_eye_center_x'],errors='coerce')
data['left_eye_center_y']= pd.to_numeric(data['left_eye_center_y'],errors='coerce')
data['right_eye_center_x']= pd.to_numeric(data['right_eye_center_x'],errors='coerce')
data['right_eye_center_y']= pd.to_numeric(data['right_eye_center_y'],errors='coerce')
data['left_eye_inner_corner_x']= pd.to_numeric(data['left_eye_inner_corner_x'],errors='coerce')
data['left_eye_inner_corner_y']= pd.to_numeric(data['left_eye_inner_corner_y'],errors='coerce')
data['left_eye_outer_corner_x']= pd.to_numeric(data['left_eye_outer_corner_x'],errors='coerce')
data['left_eye_outer_corner_y']= pd.to_numeric(data['left_eye_outer_corner_y'],errors='coerce')
data['right_eye_inner_corner_x']= pd.to_numeric(data['right_eye_inner_corner_x'],errors='coerce')
data['right_eye_inner_corner_y']= pd.to_numeric(data['right_eye_inner_corner_y'],errors='coerce')
data['right_eye_outer_corner_x']= pd.to_numeric(data['right_eye_outer_corner_x'],errors='coerce')
data['right_eye_outer_corner_y']= pd.to_numeric(data['right_eye_outer_corner_y'],errors='coerce')
data['left_eyebrow_inner_end_x']= pd.to_numeric(data['left_eyebrow_inner_end_x'],errors='coerce')
data['left_eyebrow_inner_end_y']= pd.to_numeric(data['left_eyebrow_inner_end_y'],errors='coerce')
data['left_eyebrow_outer_end_x']= pd.to_numeric(data['left_eyebrow_outer_end_x'],errors='coerce')
data['left_eyebrow_outer_end_y']= pd.to_numeric(data['left_eyebrow_outer_end_y'],errors='coerce')
data['right_eyebrow_inner_end_x']= pd.to_numeric(data['right_eyebrow_inner_end_x'],errors='coerce')
data['right_eyebrow_inner_end_y']= pd.to_numeric(data['right_eyebrow_inner_end_y'],errors='coerce')
data['right_eyebrow_outer_end_x']= pd.to_numeric(data['right_eyebrow_outer_end_x'],errors='coerce')
data['right_eyebrow_outer_end_y']= pd.to_numeric(data['right_eyebrow_outer_end_y'],errors='coerce')
data['nose_tip_x']= pd.to_numeric(data['nose_tip_x'],errors='coerce')
data['nose_tip_y']= pd.to_numeric(data['nose_tip_y'],errors='coerce')
data['mouth_left_corner_x']= pd.to_numeric(data['mouth_left_corner_x'],errors='coerce')
data['mouth_left_corner_y']= pd.to_numeric(data['mouth_left_corner_y'],errors='coerce')
data['mouth_right_corner_x']= pd.to_numeric(data['mouth_right_corner_x'],errors='coerce')
data['mouth_right_corner_y']= pd.to_numeric(data['mouth_right_corner_y'],errors='coerce')
data['mouth_center_top_lip_x']= pd.to_numeric(data['mouth_center_top_lip_x'],errors='coerce')
data['mouth_center_top_lip_y']= pd.to_numeric(data['mouth_center_top_lip_y'],errors='coerce')
data['mouth_center_bottom_lip_x']= pd.to_numeric(data['mouth_center_bottom_lip_x'],errors='coerce')
data['mouth_center_bottom_lip_y']= pd.to_numeric(data['mouth_center_bottom_lip_y'],errors='coerce')
```

```python
pd.plotting.scatter_matrix(data)
pyplot.show()
```
